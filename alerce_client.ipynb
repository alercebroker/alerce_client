```python
from alerce.core import Alerce

survey = 'ztf'

alerce = Alerce(survey)

params = {
    "survey_id": "ztf",
    "oid": "36028933560013380",
}

object_params = {
   "survey": "lsst",
   "classifier": "lc_classifier_lsst",
   "classifier_version": "2",
   "class_name":"CV/Nova"
}
```

# Object

## FORMAT JSON

```python
objects = alerce.query_objects_multisurvey(**object_params)
print(objects)
```

## FORMAT PANDAS

```python
objects = alerce.query_objects_multisurvey(format="pandas", **object_params)
print(objects)
```

## Detections

```python
detections = alerce.query_detections_multisurvey(**params)
print(detections)
```

## Non detections

```python
non_detections = alerce.query_non_detections_multisurvey(**params)
print(non_detections)
```

## Forced photometry

```python
forced_photometry = alerce.query_forced_photometry_multisurvey(**params)
print(forced_photometry)
```

## Lightcurve

```python
lightcurve = alerce.query_lightcurve_multisurvey(**params)
print(lightcurve)
```
